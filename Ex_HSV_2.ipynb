```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import sys
import os
import urllib.request
```

```python
urllib.request.urlretrieve('https://s2.glbimg.com/qdbelSRu3hawcwZlfuBmJDjrNJ0=/1200x/smart/filters:cover():strip_icc()/s.glbimg.com/po/tt2/f/original/2018/03/07/windows-xp-recebe-atualizacao-emergencial-de-seguranca.jpg', '/windows-xp-recebe-atualizacao-emergencial-de-seguranca.jpg')
```

```python
figsize = (10,10)
```

```python
bgr_img = cv2.imread('/windows-xp-recebe-atualizacao-emergencial-de-seguranca.jpg')
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=figsize)
plt.imshow(rgb_img)
plt.title('original')
plt.show()
```

```python
rgb_green = np.uint8([[[0, 255,0]]])
hsv_green = cv2.cvtColor(rgb_green, cv2.COLOR_BGR2HSV)[0,0,:]
print(hsv_green)
```

```python
hsv_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2HSV)

lower_green = hsv_green - np.array([70,200,200])
upper_green = hsv_green + np.array([30,0,0])

mask = cv2.inRange(hsv_img, lower_green, upper_green)

plt.figure(figsize=figsize)
plt.imshow(mask)
plt.title('resulted')
plt.show()
```

```python
rgb_res = cv2.bitwise_and(rgb_img,rgb_img, mask=mask)

plt.figure(figsize=figsize)
plt.imshow(rgb_res)
plt.title('output img')
plt.show()
```
