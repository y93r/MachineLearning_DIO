```python
import os
import urllib.request

# Obter a imagem e Haar Cascade
urllib.request.urlretrieve('https://static.itapemafm.com.br/s3fs-public/styles/itapema_blog_post_header/public/graphql-upload-files/metallica%20datas%20show.jpg?_Si9xhJhGHYQNsqGsoH_AyI0mVVo6oFt&itok=OKU_LEmD', './metallica%20datas%20show.jpg')
urllib.request.urlretrieve('https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_alt.xml', 
                           'haarcascade_frontalface_default.xml')

os.listdir()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Carregar o cascade
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
#Lê a imagem e converte para escala de cinza
img = cv2.imread('metallica%20datas%20show.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Faz as classificações
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

for (x, y, w, h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
# Resultado final
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
