```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import requests

url = ('https://conteudo.imguol.com.br/c/esporte/a6/2021/08/06/06082021---jogos-olimpicos-toquio-2020---volei-feminino-semifinal-brasil-x-coreia-do-sul-na-foto-as-atletas-da-selecao-comemoram-a-classificacao-para-final-foto-gaspar-nobregacob-1628271151872_v2_900x506.jpg')
r = requests.get(url)
with open ('test.jpg', 'wb') as f:
  f.write(r.content)
```

```python
img = cv2.imread('test.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.GaussianBlur(gray, (3,3),0)
```

```python
sobelx = cv2.Sobel(src=img,ddepth=cv2.CV_64F, dx=1,dy=0, ksize=5) 
sobely = cv2.Sobel(src=img,ddepth=cv2.CV_64F, dx=0,dy=1, ksize=5) 
sobelxy = cv2.Sobel(src=img,ddepth=cv2.CV_64F, dx=1,dy=1, ksize=5) 

plt.figure(figsize=(18,19))
plt.subplot(221)
plt.imshow(img, cmap = 'gray')
plt.title('original')
plt.axis('off')

plt.subplot(222)
plt.imshow(sobelxy, cmap = 'gray')
plt.title('sobelxy')
plt.axis('off')

plt.subplot(223)
plt.imshow(sobelx, cmap = 'gray')
plt.title('sobelx')
plt.axis('off')

plt.subplot(224)
plt.imshow(sobely, cmap = 'gray')
plt.title('sobely')
plt.axis('off')
```
