```python
import time
import cv2
import numpy as np
import urllib.request

# Carregar imagem da internet
resp = urllib.request.urlopen('http://rededoesporte.gov.br/pt-br/noticias/prata-no-volei-feminino-fecha-participacao-do-brasil-em-toquio-com-201cchave-de-ouro201d/08082021_volei_32.jpg/image')
imagem = np.asarray(bytearray(resp.read()), dtype="uint8")
imagem = cv2.imdecode(imagem, cv2.IMREAD_COLOR)

# Carregar imagem do PC
# imagem = cv2.imread("volei_32.jpg")
```

```python
# FOR LOOPS

start_time = time.time()

for coluna in imagem:
    for pixel in coluna:
        media = int(sum(pixel)/3)
        pixel [0] = media
        pixel [1] = media
        pixel [2] = media

cv2.imwrite("http://rededoesporte.gov.br/pt-br/noticias/prata-no-volei-feminino-fecha-participacao-do-brasil-em-toquio-com-201cchave-de-ouro201d/08082021_volei_32.jpg/image",imagem)
print("--- %s segundos ---" % (time.time() - start_time))
```

```python
from google.colab.patches import cv2_imshow
# MATRIZES E ARRAYS

# Carregar imagem da internet
resp = urllib.request.urlopen("http://rededoesporte.gov.br/pt-br/noticias/prata-no-volei-feminino-fecha-participacao-do-brasil-em-toquio-com-201cchave-de-ouro201d/08082021_volei_32.jpg/image")
imagem = np.asarray(bytearray(resp.read()), dtype="uint8")
imagem = cv2.imdecode(imagem, cv2.IMREAD_COLOR)

# Carregar imagem do PC
# imagem = cv2.imread("volei_32.jpg")

start_time = time.time()

imagem = 0.0722 * imagem[:, :, 0] + 0.7152 * imagem[:, :, 1] + 0.2126 * imagem[:, :, 2]
imagem = imagem.astype(np.uint8)

cv2.imwrite("teste de cinza.jpg",imagem)
cv2_imshow(imagem)
print("--- %s segundos ---" % (time.time() - start_time))
```
