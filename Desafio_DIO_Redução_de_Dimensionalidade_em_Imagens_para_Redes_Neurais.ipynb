```python
import IPython.display as display
from PIL import Image
display.display(Image.open('guitarra.png'))
```

```python
# abre a imagem colorida
img_colorida = Image.open('guitarra.png')

# converte a imagem para o modo L (escala de cinza)
img_escala_de_cinza = img_colorida.convert('L')

# salva a nova imagem
img_escala_de_cinza.save('guitarra_cinza.png')
display.display(Image.open('guitarra_cinza.png'))
```
